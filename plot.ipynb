```python
import numpy as np
import matplotlib.pyplot as pyplot 

data = np.loadtxt("cifar10_training.txt")

epoch = data[:, 0]
train_loss = data[:, 1]
evaluate_loss = data[:, 2]
acc = data[:, 3]


pyplot.plot(epoch, acc, color = 'navy', label="Evaluation Accuracy")
pyplot.plot(epoch, acc, 'P', color = 'coral')
pyplot.xlabel("Epoch")
pyplot.ylabel("Accuracy")
pyplot.legend()
pyplot.title("Accuracy for CIFAR10")
# pyplot.savefig("./images/cifar10_acc.png", dpi = 200)
pyplot.show()
```

```python
pyplot.plot(epoch, train_loss, color = 'navy', label="Training loss")
pyplot.plot(epoch, train_loss, 'P', color = 'dodgerblue')
pyplot.plot(epoch, evaluate_loss, color = 'darkred', label="Evaluation loss")
pyplot.plot(epoch, evaluate_loss, 'P', color = 'lightcoral')
pyplot.xlabel("Epoch")
pyplot.ylabel("Loss")
pyplot.legend()
pyplot.title("Loss for CIFAR10")
# pyplot.savefig("./images/cifar10_loss.png", dpi = 200)
pyplot.show()
```

```python
data = np.loadtxt("food101_training.txt")

epoch = data[:, 0]
train_loss = data[:, 1]
evaluate_loss = data[:, 2]
top1_acc = data[:, 3]
top5_acc = data[:, 4]

pyplot.plot(epoch, top1_acc, color = 'navy', label="Evaluation TOP1 Accuracy")
pyplot.plot(epoch, top1_acc, 'P', color = 'dodgerblue')
pyplot.plot(epoch, top5_acc, color = 'darkred', label="Evaluation TOP5 Accuracy")
pyplot.plot(epoch, top5_acc, '^', color = 'lightcoral')

pyplot.xlabel("Epoch")
pyplot.ylabel("Accuracy")
pyplot.legend()
pyplot.title("Accuracy for Food101")
pyplot.savefig("./images/food101_acc.png", dpi = 200)
pyplot.show()




```

```python
pyplot.plot(epoch, train_loss, color = 'navy', label="Training Loss")
pyplot.plot(epoch, train_loss, 'P', color = 'dodgerblue')
pyplot.plot(epoch, evaluate_loss, color = 'darkred', label="Evaluation Loss")
pyplot.plot(epoch, evaluate_loss, '^', color = 'lightcoral')

pyplot.xlabel("Epoch")
pyplot.ylabel("Loss")
pyplot.legend()
pyplot.title("Loss for Food101")
pyplot.savefig("./images/food101_loss.png", dpi = 200)
pyplot.show()
```
